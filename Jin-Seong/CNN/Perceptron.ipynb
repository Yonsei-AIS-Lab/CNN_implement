```python
import os
import random
import numpy as np

import matplotlib.pyplot as plt
```

```python
x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])
```

```python
def step(input_value):
    if input_value > 0:
        result = 1
    else:
        result = -1
    return result

def sigmoid(z):
    return 1 / (1+np.exp(-z))
```

```python
def MSE(y, y_hat):
    return np.mean((y-y_hat)**2)
```

```python
class Perceptron:
    def __init__(self, input_size, activation):
        self.weights = np.random.random(input_size)
        self.bias = random.random()
        self.activation = activation
        self.history = []

    def printHistory(self):
        history = np.array(self.history)
        plt.figure()
        plt.xlabel('Epochs')
        plt.ylabel('Weights')
        plt.title('Weights per Epoch')

        plt.plot(history[:, 0], label='w1')
        plt.plot(history[:, 1], label='w2')
        plt.legend()
        plt.show()        

    def feedForward(self, input_x):      
        y_hat = input_x.dot(self.weights)
        return y_hat
    
    def predict(self, input_x):
        y_hat = self.feedForward(input_x)
        if y_hat > 0.5:
            result = 1
        else:
            result = 0

        return result

    def training(self, train_x, train_y, learning_rate, epochs):
        for i in range(epochs):
            #print('{0} epochs'.format(i))
            self.history.append(self.weights)
            y_hat = self.feedForward(train_x)
            diff_y = np.mean(train_y - y_hat)
            self.weights = self.weights + diff_y*learning_rate
```

```python
single_perceptron = Perceptron(2, sigmoid)
```

```python
single_perceptron.training(x, y, 0.001, 10000)
```

```python
single_perceptron.printHistory()
```

```python
test_x = np.array([1, 0])
single_perceptron.predict(test_x)
```
